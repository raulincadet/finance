```python
table=soup.find_all('table',class_="sc-6af360b8-1 fzhtim")

data_rows = []

# Extract data from the table
for row in table[0].find_all('tr'):
    # Initialize list to store cell data for this row
    row_data = []
    # Iterate over cells in the row
    for cell in row.find_all(['td', 'th']):  # Include th for header cells if needed
        # Extract text from the cell and append to row_data
        row_data.append(cell.get_text(strip=True))
    # Append row_data to data_rows
    data_rows.append(row_data)

# Assuming the first row contains headers, create a DataFrame
data = pd.DataFrame(data_rows[1:], columns=data_rows[0])

# Display the data
print(data)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the website
url = 'https://www.trademap.org/Country_SelProductCountry_TS.aspx?nvpm=1||||080550'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the data (assuming it's in a table)
table = soup.find('table', id='ctl00_PageContent_GridViewPanel1')

# Initialize lists to store data
exporters = []
values = []

# Extract data from the table
for row in table.find_all('tr')[1:]:
    cols = row.find_all('td')
    exporters.append(cols[0].text.strip())
    values.append(cols[1].text.strip())

# Create a DataFrame from the extracted data
data = pd.DataFrame({'Exporter': exporters, 'Value': values})

# Display the data
print(data)
```
